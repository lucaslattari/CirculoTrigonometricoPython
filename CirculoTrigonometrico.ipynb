```
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```
plt.style.use("default")

def plotUnitCircle(x, y):
  #cria plot
  axes = plt.subplot()
  axes.grid(True)

  #cria círculo trigonométrico
  c = plt.Circle((0, 0), radius = 1.0, edgecolor = 'black', facecolor = 'none')
  axes.add_artist(c)

  #configura eixos
  axes.axhline(y = 0, color = 'black')
  axes.axvline(x = 0, color = 'black')

  #define triângulo
  axes.plot([0, x], [0, 0], color = 'blue', label = 'cateto adjacente')
  axes.plot([x, x], [0, y], color = 'green', label = 'cateto oposto')
  axes.plot([0, x], [0, y], color = 'red', label = 'hipotenusa')

  #limites do meu gráfico
  if x > 0:
    plt.xlim(-0.1, 1.1)
  else:
    plt.xlim(-1.1, 0.1)
  if y > 0:
    plt.ylim(-0.1, 1.1)
  else:
    plt.ylim(-1.1, 0.1)

  #renderiza gráfico
  plt.legend()
  plt.title("Círculo Trigonométrico")
  plt.show()
```

# Conversão de Graus para Radianos

Na conta abaixo, *rad* é o valor desejado em radianos e *g* é o ângulo em graus.

\begin{equation}
180  -\!\!\!-\!\!\!- \pi \\
g  -\!\!\!-\!\!\!- rad
\end{equation}

$ 180 \times rad = g \times \pi $ \\
$ rad = \frac{g \times \pi}{180} $

# Renderização de Círculo Trigonométrico Unitário

## 30º

```
d = 30

angle_radians = math.radians(d)
x = math.cos(angle_radians)
y = math.sin(angle_radians)

print("x =", x)
print("y =", y)

plotUnitCircle(x, y)
```
